```python
for x in range(1,6):
    #fort loop with five interations - pages 1-6
    html = browser.html
    #create HTML object assigned to html variable
    soup = BeautifulSoup(html, 'html.parser')
    #use BS4() to parse html object 
    quotes = soup.find_all('span', class_='text')
    # use soup to find all span tags that are of the text class 
    for quote in quotes:
        #print statements wrapped in another for loop that will print 
        #everything that soup found that has span and class text
        print('page:',x,'----------')
        print(quote.text)
    browser.links.find_by_partial_text('Next')
```
